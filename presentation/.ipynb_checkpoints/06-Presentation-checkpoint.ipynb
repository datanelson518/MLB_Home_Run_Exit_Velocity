### Total Home Runs By Year

```python
hr_totals = presentation_df.groupby(['game_year']).agg({'pitch_type':'count'})
hr_totals.columns = ['home_run_total']
```

```python
plt.figure(figsize=(25, 12))

colors = ["light green", "lightish blue", "orange red"]
palette=sns.xkcd_palette(colors)
sns.set_context('poster', rc={'patch.linewidth':2})

plt.title('Home Runs by Season', fontsize=30, pad=60)

a = sns.barplot(hr_totals.index,
                hr_totals['home_run_total'],
                palette=palette,
                edgecolor='k',
                alpha=.5)

sns.despine()

plt.xlabel('MLB Season', fontsize=25, labelpad=40)
plt.ylabel('Home Runs', fontsize=25, labelpad=40)

greenpatches = mpatches.Patch(edgecolor='k', facecolor='#96f97b', label='2015', alpha=.5)
bluepatches = mpatches.Patch(edgecolor='k', facecolor='#3d7afd', label='2016', alpha=.5)
redpatches = mpatches.Patch(edgecolor='k', facecolor='#fd411e', label='2017', alpha=.5)

plt.legend(title='MLB Season', handles=[greenpatches, bluepatches, redpatches], loc=1, fontsize=25)

plt.tight_layout();
plt.savefig('../presentation/hr_seasons.png')
```

### Distrubution of Data

```python
plt.figure(figsize=(30, 18))

colors = ["light green", "lightish blue", "orange red"]
palette=sns.xkcd_palette(colors)
sns.set_context('poster')

plt.title('Launch Speed (2015 - 2017)', fontsize=40, pad=60)
a = sns.distplot(presentation_df['launch_speed'], 
                 kde_kws={"color":"k","lw":4,"label":"Launch Speed Estim"},
                 hist_kws={'edgecolor':"k", 'linewidth':2, 'color':'#fd411e'},
                 label='Launch Speed Dist')

plt.xlabel('Speed (MPH)', fontsize=30,labelpad=40)
plt.ylabel('Distribution (%)', fontsize=30, labelpad=40)

leg = plt.legend(fontsize=30, loc=0)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(1)

a.tick_params(labelsize=30,labelcolor="k")

plt.tight_layout();
plt.savefig('../presentation/velo_distribution.png')
```

### Launch Speed and Angles

```python
presentation_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
speed_angle = presentation_df.sort_values('launch_speed', ascending=False)[['launch_speed', 'launch_angle']]
```

```python
ps = []
color=[]
for x in speed_angle['launch_speed']:
    if x >= 106:
        ps.append(4000)
        color.append('#96f97b')
    elif 100 <= x < 106:
        ps.append(2000)
        color.append('#3d7afd')
    elif x < 100:
        ps.append(1000)
        color.append('#fd411e')
speed_angle['point_size'] = ps
speed_angle['color'] = color
```

```python
plt.figure(figsize=(30, 18))

sns.set_context('poster')

plt.title('Launch Speeds & Angles (2015 - 2017)', fontsize=40, pad=60)

a = sns.regplot(speed_angle['launch_angle'], 
                speed_angle['launch_speed'],
                scatter_kws={'s':speed_angle['point_size'], 
                             'facecolors':speed_angle['color'], 
                             'alpha':.5, 
                             'edgecolor':'k'},
                fit_reg=False)


plt.xlabel('Launch Angle (Degrees)', fontsize=30, labelpad=40)
plt.ylabel('Launch Speed (MPH)', fontsize=30, labelpad=40)


a.tick_params(labelsize=30,labelcolor="k")

sns.despine()

greenpatches = mpatches.Patch(edgecolor='k', facecolor='#96f97b', label='106+ MPH', alpha=.5)
bluepatches = mpatches.Patch(edgecolor='k', facecolor='#3d7afd', label='100 MPH - 106 MPH', alpha=.5)
redpatches = mpatches.Patch(edgecolor='k', facecolor='#fd411e', label='100- MPH', alpha=.5)

leg = plt.legend(title='Launch Speeds', handles=[greenpatches, bluepatches, redpatches], loc=2, fontsize=30)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(1)

plt.tight_layout();
plt.savefig('../presentation/launch_speeds_angles.png')
```

### Launch Speed With Distance

```python
hr_distance = presentation_df[['launch_speed', 'hit_distance_sc']].copy()
hr_distance.sort_values('hit_distance_sc', ascending=False, inplace=True)
```

```python
ps = []
color=[]
for x in hr_distance['hit_distance_sc']:
    if x >= 450:
        ps.append(3000)
        color.append('#96f97b')
    elif 400 <= x < 450:
        ps.append(1000)
        color.append('#3d7afd')
    elif x < 400:
        ps.append(500)
        color.append('#fd411e')
hr_distance['point_size'] = ps
hr_distance['color'] = color
```

```python
plt.figure(figsize=(30, 18))

sns.set_context('poster', rc={'patch.linewidth':2})

plt.title('Launch Speeds and Distance (2015 - 2017)', fontsize=40, pad=60)

a = sns.regplot(hr_distance['launch_speed'], 
                hr_distance['hit_distance_sc'],
                scatter_kws={'s':hr_distance['point_size'], 
                             'facecolors':hr_distance['color'], 
                             'alpha':.5,
                             'edgecolor':'k'},
                fit_reg=False)


plt.xlabel('Launch Speed (MPH)', fontsize=30, labelpad=40)
plt.ylabel('Distance (Feet)', fontsize=30, labelpad=40)


a.tick_params(labelsize=30, labelcolor="k")
sns.despine()

greenpatches = mpatches.Patch(edgecolor='k', facecolor='#96f97b', label='450+ Feet', alpha=.5)
bluepatches = mpatches.Patch(edgecolor='k', facecolor='#3d7afd', label='400 Feet - 450 Feet', alpha=.5)
redpatches = mpatches.Patch(edgecolor='k', facecolor='#fd411e', label='400- Feet', alpha=.5)

leg = plt.legend(title='Home Run Distance', handles=[greenpatches, bluepatches, redpatches], loc=2, fontsize=30)
leg.get_frame().set_edgecolor('k')
leg.get_frame().set_linewidth(1)

plt.tight_layout();
plt.savefig('../presentation/launch_speeds_distance.png')
```

### Home Runs by Pitch

The below plot will provide insight into MLBs total number of home runs by pitch type.

- `pitch_type`: The type of pitch that was thrown and hit for a home run
    - `CH` : `Changeup`
    - `CU` : `Curveball`
    - `EP` : `Eephus`
    - `FC` : `Cut Fastball (Cutter)`
    - `FF` : `Four-seam Fastball`
    - `FO` : `Pitch Out`
    - `FS` : `Sinking Fastball / Split-Fingered (Splitter)`
    - `FT` : `Two-seam Fastball`
    - `KC` : `Knuckle-curve`
    - `KN` : `Knuckleball`
    - `SC` : `Screwball`
    - `SI` : `Sinker`
    - `SL` : `Slider`

```python
hr_pitch = presentation_df.groupby(['game_year', 'pitch_type']).agg({'pitch_type':'count'})
hr_pitch.columns = ['home_run_total']
```

```python
hr_pitch.reset_index(inplace=True)
```

```python
hr_pitch.sort_values('home_run_total', inplace=True, ascending=False)
```

```python
plt.figure(figsize=(30, 18))

colors = ["light green", "lightish blue", "orange red"]
palette=sns.xkcd_palette(colors)
sns.set_context('poster', rc={'patch.linewidth':2})

plt.title('Home Runs by Pitch', fontsize=40, pad=60)

a = sns.barplot(hr_pitch['pitch_type'],
                hr_pitch['home_run_total'],
                hue=hr_pitch['game_year'],
                palette=palette,
                edgecolor='k',
                alpha=.5)

sns.despine()

plt.xlabel('Pitch Type', fontsize=35, labelpad=40)
plt.ylabel('Home Runs', fontsize=35, labelpad=40)

a.tick_params(labelsize=35, labelcolor="k")

greenpatches = mpatches.Patch(edgecolor='k', facecolor='#96f97b', label='2015', alpha=.5)
bluepatches = mpatches.Patch(edgecolor='k', facecolor='#3d7afd', label='2016', alpha=.5)
redpatches = mpatches.Patch(edgecolor='k', facecolor='#fd411e', label='2017', alpha=.5)

plt.legend(title='MLB Season', handles=[greenpatches, bluepatches, redpatches], loc=1, fontsize=30)

plt.tight_layout();
plt.savefig('../presentation/home_runs_pitch.png')
```
